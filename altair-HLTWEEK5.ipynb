```python
import altair as alt
import pandas as pd
```

```python
ad= pd.read_csv(r"/Users/rania/Downloads/Automobile_data.csv")
ad.head()
```

```python
brush = alt.selection_interval()
```

```python
points = alt.Chart(ad).mark_point().encode(
    x='company:N',
    y='price:Q').add_selection(brush)

points
```

```python
alt.Chart(ad).mark_point().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='Origin:N'
).properties(
    width=100,
    height=100
).repeat(
    row=['price', 'average_mileage', 'horsepower'],
    column=['average_mileage', 'Horsepower','price']
).interactive()
```

```python
bars = alt.Chart(ad).mark_bar().encode(
    y='company:N',
    color='price:Q',
    x='price:Q'
)

bars
```

```python
import altair as alt
import numpy as np
import pandas as pd


alt.Chart(ad,width=300, height=300).mark_line().encode(
    y="price:Q",
    x="engine-type:N"
)
```

```python
alt.Chart(ad, width=300, height=300).mark_bar().encode(
    x=alt.X('price:Q', bin=alt.Bin(maxbins=10)),
    y='engine-type:N')

```

```python
alt.Chart(ad).mark_bar().encode(
    alt.X("price:Q", bin=True),
    y='count()',
)
```

```python
alt.Chart(ad).mark_circle(size=60).encode(
    x='Horsepower:Q',
    y='price:Q',
    color='company',
    tooltip=[ 'company', 'Horsepower', 'price']
).interactive()
```
